```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/MACHINE_LEARNING_PROJECTs/CarPrice_Assignment.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['CarName', 'fueltype', 'aspiration', 'doornumber', 'carbody','drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']
# Apply Label Encoding to each column
en= LabelEncoder()
for col in categorical_cols:
    df[col] = en.fit_transform(df[col])
df
```

```python
x=df.iloc[:,1:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
m=model.coef_
m
```

```python
c=model.intercept_
c
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
print("Mean Absolute Error=",mae)
print("Mean Squared Error=",mse)
print("root mean square error:",(mse**0.5))
```
